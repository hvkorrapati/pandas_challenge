```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../04-Pandas/Instructions/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "../04-Pandas/Instructions/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# # District Summary
# Calculate the total number of schools

# Calculate the total number of students

# Calculate the total budget

# Calculate the average math score

# Calculate the average reading score

# Calculate the percentage of students with a passing math score (70 or greater)

# Calculate the percentage of students with a passing reading score (70 or greater)

# Calculate the percentage of students who passed math and reading (% Overall Passing)

# Create a dataframe to hold the above results

# Optional: give the displayed data cleaner formatting


total_number_schools = len(pd.unique(school_data_complete["school_name"]))
total_number_students = len(pd.unique(school_data_complete['student_name']))
total_budget = pd.unique(school_data_complete['budget']).sum()
avg_math = school_data_complete['math_score'].mean()
avg_reading = school_data_complete['reading_score'].mean()

passing_math = school_data_complete.loc[(school_data_complete["math_score"] >= 70),:]
pct_passing_math = len(passing_math)/len(school_data_complete["math_score"])
school_data_complete['passing math'] = school_data_complete['math_score']>=70

passing_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 70),:]
pct_passing_reading = len(passing_reading)/len(school_data_complete["reading_score"])
school_data_complete['passing reading'] = school_data_complete['reading_score']>=70

passing_both = school_data_complete.loc[((school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)),:]
pct_passing_both = len(passing_both)/len(school_data_complete)
school_data_complete['passing both'] = (school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)

district_summary = pd.DataFrame({'total schools': [total_number_schools],
                                 'total students': [total_number_students],
                                 'total budget': [total_budget],
                                 'average math score': [avg_math],
                                 'percent passing math': [pct_passing_math],
                                 'average reading score': [avg_reading],
                                 'percent passing reading': [pct_passing_reading],
                                 'percent passing math and reading': [pct_passing_both]
                                 
                                })
district_summary
```

```python
# School Summary
# Create an overview table that summarizes key metrics about each school, including:

# School Name
# School Type
# Total Students
# Total School Budget
# Per Student Budget
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# % Overall Passing (The percentage of students that passed math and reading.)
# Create a dataframe to hold the above results

school_summary = school_data_complete.groupby(['school_name'])

school_name = school_summary['school_name'].unique()
school_type = school_summary['type'].unique()
school_total = school_summary['student_name'].count()
school_budget = school_summary['budget'].unique()
school_budget_per_student = school_budget/school_total
school_avg_math = school_summary['math_score'].mean()
school_avg_reading = school_summary['reading_score'].mean()

school_passing_math = ((school_summary['passing math']).sum())/school_total
school_passing_reading = ((school_summary['passing reading']).sum())/school_total
school_passing_both = ((school_summary['passing both']).sum())/school_total

school_summary_df = pd.DataFrame({
                                    'school name': school_name,
                                 'school type': school_type,
                                    'total school budget': school_budget,
                                 'per student budget': school_budget_per_student,
                                 'average math score': school_avg_math,
                                 'percent passing math': school_passing_math,
                                 'average reading score': avg_reading,
                                 'percent passing reading': school_passing_reading,
                                 'percent overall passing': school_passing_both
                                 
                                })

school_summary_df
```

```python
# Top Performing Schools (By % Overall Passing)
# Sort and display the top five performing schools by % overall passing.

school_summary_df.sort_values(by = ['percent overall passing'], ascending = False).head(5)
```

```python
# Bottom Performing Schools (By % Overall Passing)
# Sort and display the five worst-performing schools by % overall passing.

school_summary_df.sort_values(by = ['percent overall passing'], ascending = True).head(5)
```

```python
# Math Scores by Grade
# Create a table that lists the average Reading Score for students of each grade level 
# (9th, 10th, 11th, 12th) at each school.

# Create a pandas series for each grade. Hint: use a conditional statement.

# Group each series by school

# Combine the series into a dataframe

# Optional: give the displayed data cleaner formatting

ninth_math_grade = []
tenth_math_grade = []
eleventh_math_grade = []
twelfth_math_grade = []

for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '9th':
        ninth_math = school_data_complete.iloc[i,6]
    else:
        ninth_math = float('Nan') 
    ninth_math_grade.append(ninth_math)


for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '10th':
        tenth_math = school_data_complete.iloc[i,6] 
    else:
        tenth_math = float('Nan') 
    tenth_math_grade.append(tenth_math)

    
for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '11th':
        eleventh_math = school_data_complete.iloc[i,6]
    else:
        eleventh_math = float('Nan') 
    eleventh_math_grade.append(eleventh_math)
    
    
for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '12th':
        twelfth_math = school_data_complete.iloc[i,6]
    else:
        twelfth_math = float('Nan')  
    twelfth_math_grade.append(twelfth_math)

school_data_complete['ninth math grade'] = ninth_math_grade
school_data_complete['tenth math grade'] = tenth_math_grade
school_data_complete['eleventh math grade'] = eleventh_math_grade
school_data_complete['twelfth math grade'] = twelfth_math_grade
        

math_by_class = school_data_complete.groupby(['school_name'])

ninth_average_math = math_by_class['ninth math grade'].mean()
tenth_average_math = math_by_class['tenth math grade'].mean()
eleventh_average_math = math_by_class['eleventh math grade'].mean()
twelfth_average_math = math_by_class['twelfth math grade'].mean()

math_by_class_df = pd.DataFrame({'9th grade average math score': ninth_average_math,
                                '10th grade average math score': tenth_average_math,
                                '11th grade average math score': eleventh_average_math,
                                '12th grade average math score': twelfth_average_math,
                                })

math_by_class_df
```

```python
# Reading Score by Grade
# Perform the same operations as above for reading scores

ninth_reading_grade = []
tenth_reading_grade = []
eleventh_reading_grade = []
twelfth_reading_grade = []

for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '9th':
        ninth_reading = school_data_complete.iloc[i,5]
    else:
        ninth_reading = float('Nan') 
    ninth_reading_grade.append(ninth_reading)


for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '10th':
        tenth_reading = school_data_complete.iloc[i,5] 
    else:
        tenth_reading = float('Nan') 
    tenth_reading_grade.append(tenth_reading)

    
for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '11th':
        eleventh_reading = school_data_complete.iloc[i,5]
    else:
        eleventh_reading = float('Nan') 
    eleventh_reading_grade.append(eleventh_reading)
    
    
for i in range(len(school_data_complete)):    
    if school_data_complete.iloc[i, 3] == '12th':
        twelfth_reading = school_data_complete.iloc[i,5]
    else:
        twelfth_reading = float('Nan')  
    twelfth_reading_grade.append(twelfth_reading)

school_data_complete['ninth reading grade'] = ninth_reading_grade
school_data_complete['tenth reading grade'] = tenth_reading_grade
school_data_complete['eleventh reading grade'] = eleventh_reading_grade
school_data_complete['twelfth reading grade'] = twelfth_reading_grade
        

reading_by_class = school_data_complete.groupby(['school_name'])

ninth_average_reading = reading_by_class['ninth reading grade'].mean()
tenth_average_reading = reading_by_class['tenth reading grade'].mean()
eleventh_average_reading = reading_by_class['eleventh reading grade'].mean()
twelfth_average_reading = reading_by_class['twelfth reading grade'].mean()

reading_by_class_df = pd.DataFrame({'9th grade average reading score': ninth_average_reading,
                                '10th grade average reading score': tenth_average_reading,
                                '11th grade average reading score': eleventh_average_reading,
                                '12th grade average reading score': twelfth_average_reading,
                                })

reading_by_class_df
```

```python
# Scores by School Spending
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
# Use 4 reasonable bins to group school spending. Include in the table each of the following:
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)

budget_per_student = []

for i in range(len(school_data_complete)):
    budget_var = int(school_budget_per_student[school_data_complete.iloc[i,4]])
    budget_per_student.append(budget_var)
    
    
school_data_complete['budget per student'] = budget_per_student

spending_bins = [0,585,630,645,680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

spending_range = pd.cut(school_data_complete['budget per student'],bins = spending_bins,labels = spending_labels)

school_data_complete['per student spending range'] = spending_range
per_student_summary = school_data_complete.groupby(['per student spending range'])


spending_school_total = per_student_summary['student_name'].count()
spending_avg_math = per_student_summary['math_score'].mean()
spending_avg_reading = per_student_summary['reading_score'].mean()

spending_passing_math = ((per_student_summary['passing math']).sum())/spending_school_total
spending_passing_reading = ((per_student_summary['passing reading']).sum())/spending_school_total
spending_passing_both = ((per_student_summary['passing both']).sum())/spending_school_total

per_student_summary_df = pd.DataFrame({'Avg math score': spending_avg_math,
                                      'Avg reading score': spending_avg_reading,
                                       'Pct passing math': spending_passing_math,
                                       'Pct passing reading': spending_passing_reading,
                                       'Pct passing both': spending_passing_both
                                      })

per_student_summary_df
```

```python
# Scores by School Size
# Perform the same operations as above, based on school size.

size_bins = [0,1000,2000,5000]
size_labels = ["Small", "Medium", "Large"]

size_range = pd.cut(school_data_complete['size'],bins = size_bins,labels = size_labels)

school_data_complete['size range'] = size_range
size_summary = school_data_complete.groupby(['size range'])


size_school_total = size_summary['student_name'].count()
size_avg_math = size_summary['math_score'].mean()
size_avg_reading = size_summary['reading_score'].mean()

size_passing_math = ((size_summary['passing math']).sum())/size_school_total
size_passing_reading = ((size_summary['passing reading']).sum())/size_school_total
size_passing_both = ((size_summary['passing both']).sum())/size_school_total

size_summary_df = pd.DataFrame({'Avg math score': size_avg_math,
                                      'Avg reading score': size_avg_reading,
                                       'Pct passing math': size_passing_math,
                                       'Pct passing reading': size_passing_reading,
                                       'Pct passing both': size_passing_both
                                      })

size_summary_df
```
